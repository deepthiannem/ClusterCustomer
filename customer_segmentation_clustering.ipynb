```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Customers.csv")
df
```

```python
df.describe()
```

# Univariate Analysis

```python
sns.distplot(df['Annual Income (k$)'],color = 'green');
```

```python
df.columns
```

```python
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.distplot(df[i],color = 'green' )
```

```python
df['Gender'] = df['Gender'].astype('category')
palette = {'Male': 'red', 'Female': 'skyblue'}
sns.kdeplot(data=df, x='Annual Income (k$)', shade=True, hue='Gender', palette=palette)
```

```python

columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
palette = {'Male': 'red', 'Female': 'skyblue'}

for i in columns:
    plt.figure()
    sns.kdeplot(data=df, x=i, shade=True, hue='Gender',palette=palette)
```

```python
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
for i in columns:
    plt.figure()
    sns.boxplot(data=df,x='Gender',y=df[i])
```
